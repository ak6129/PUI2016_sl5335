```python
import numpy as np
import datetime
import pandas as pd
import pylab as pl
import seaborn
import csv

% pylab inline
```

```python
deadlift = open('BarbellDeadlift.txt')
```

```python
deadlift_df = pd.read_csv(deadlift)
```

```python
deadlift_df.rename(columns = {'Date (YYYYMMDD)': 'Date', 'Unnamed: 3': 'Weight', 'Unnamed: 5': 'Reps'}, inplace = True)
```

```python
deadlift_df.dtypes
```

```python
deadlift_df.drop(['Set', 'unit', 'unit.1', 'Combined'], axis = 1, inplace = True)
```

```python
deadlift_df.Date = pd.to_datetime(deadlift_df.Date)
deadlift_df = deadlift_df[deadlift_df.Date > '2015-12-31'] # limits to 2016
deadlift_df.head()
```

```python
deadlift_df['Total_Weight'] = deadlift_df.Weight * deadlift_df.Reps
```

```python
deadlift_df.head()
```

```python
grouped = deadlift_df.groupby('Date')
```

```python
set_nos = []
for i, group in grouped:
    additions = range(1, len(group.index) + 1)
    set_nos.extend(additions)
    
```

```python
deadlift_df['Set'] = set_nos
```

```python
pivoted = deadlift_df.pivot_table(deadlift_df, index='Date', columns='Set', aggfunc='mean', fill_value=0)
```

```python
pl.rcParams.update({'font.size': 20})
pivoted.Total_Weight.plot(kind = 'bar', stacked = True, figsize = (20, 20), cmap = pl.cm.gist_earth, fontsize = 20)
pl.title('Total Deadlifting Volume in Pounds, 2016', fontsize = 30)
pl.ylabel('Total Weight in Pounds (Reps * Weight)', fontsize = 25)
pl.xlabel('Date', fontsize = 25)
pl.legend(title = 'Set No.', fontsize = 20)
pl.xticks(rotation = 85)
pl.show()
```
